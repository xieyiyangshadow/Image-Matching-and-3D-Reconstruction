# 01 - Image Retrieval & Shortlist Generation (Per-Dataset)

## Overview
- Process each dataset independently (matching champion's approach)
- Extract global features from 4 models: MASt3R-ASMK, MASt3R-SPoC, DINOv2, ISC
- **ASMK**: FPS-based pair extraction (n=10 seeds, k=25 per seed)
- **SPoC, DINOv2, ISC**: Top-K neighbors (k=10 each)
- Union of all 4 shortlists → per-dataset candidate pairs

**Reference**: 1st place IMC 2025 solution (per-dataset, FPS-based ASMK, MAC-pooled DINOv2)

## 1. Environment & Config

```python
import os, sys, json, time, pickle
from pathlib import Path
from collections import defaultdict
import numpy as np
import torch

import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image
from tqdm import tqdm
from matplotlib import pyplot as plt

os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

# Paths
IMAGE_ROOT = Path('../image-matching-challenge-2025/train')
OUTPUT_DIR = Path('../output/retrieval')
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

CHECKPOINT_DIR = Path('../checkpoints')
MODEL_DIR = Path('../models')

# Per-model top-K (1st place writeup parameters)
# ASMK: n=10 FPS seeds, k=25 per seed (FPS-based pair extraction)
# SPoC: topk=10, DINO: topk=10, ISC: topk=10
PER_MODEL_TOPK = {'asmk': 25, 'spoc': 10, 'dino': 10, 'isc': 10}
ASMK_FPS_N = 10  # Number of FPS seed images for ASMK

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

## 2. Collect Image Paths

```python
# Discover datasets (each subdirectory in train/ is a separate dataset)
IMAGE_EXTS = {'.jpg', '.jpeg', '.png', '.bmp', '.webp'}

datasets = {}
for d in sorted(IMAGE_ROOT.iterdir()):
    if d.is_dir() and d.name != 'outliers':
        images = sorted(p for p in d.rglob('*') if p.suffix.lower() in IMAGE_EXTS)
        if images:
            datasets[d.name] = images

print(f'Datasets: {len(datasets)}')
for ds_name, ds_images in datasets.items():
    print(f'  {ds_name}: {len(ds_images)} images')

# Save per-dataset path mappings
for ds_name, ds_images in datasets.items():
    ds_out = OUTPUT_DIR / ds_name
    ds_out.mkdir(parents=True, exist_ok=True)
    path_mapping = {p.name: str(p) for p in ds_images}
    with open(ds_out / 'image_paths.json', 'w') as f:
        json.dump(path_mapping, f, indent=2)

print(f'\nPer-dataset mappings saved to {OUTPUT_DIR}/<dataset>/image_paths.json')
```

## 3. Load Pretrained Models

### 3a. DINOv2 (HuggingFace pretrained, no checkpoint needed)
Source: `facebook/dinov2-base` → 256-dim global features

```python
from transformers import AutoModel

DINO_INPUT = 224
DINO_DIM = 768  # Raw DINOv2-base feature dimension (no projector)

class DINOv2Global(nn.Module):
    """DINOv2 global descriptor via MAC pooling (matches champion's approach).
    Uses max-pooling over patch tokens, no linear projector."""
    def __init__(self):
        super().__init__()
        self.encoder = AutoModel.from_pretrained(str(MODEL_DIR / 'dinov2-base'), local_files_only=True)
        # No projector — use raw features as the champion does
    def forward(self, x):
        outputs = self.encoder(x)
        # MAC pooling: max over patch tokens (skip CLS token at position 0)
        return F.normalize(outputs.last_hidden_state[:, 1:].max(dim=1)[0], dim=1, p=2)

dino_model = DINOv2Global().to(device).eval()
print(f'DINOv2 loaded (pretrained, MAC pooling). Dim: {DINO_DIM}')
```

### 3b. MAST3R-ASMK (trained head on MAST3R backbone)
Backbone: MAST3R ViT-Large (1024-dim encoder)
Head: ASMK codebook (4096 codewords, 1024-dim each) → 4096-dim global descriptor

```python
sys.path.insert(0, str(Path('../mast3r')))
from mast3r.model import AsymmetricMASt3R

ASMK_IMG = 224
ASMK_DIM = 4096
MAST3R_ENC_DIM = 1024  # ViT-Large
MAST3R_PATH = str(MODEL_DIR / 'MASt3R_ViTLarge_BaseDecoder_512_catmlpdpt_metric.pth')
ASMK_CKPT = str(CHECKPOINT_DIR / 'mast3r_asmk' / 'best_model.pth')
_TLK = {'weights_only': False, 'map_location': 'cpu'}

class MAST3RBackbone(nn.Module):
    def __init__(self, path):
        super().__init__()
        m = AsymmetricMASt3R.from_pretrained(path)
        self.patch_embed, self.enc_blocks, self.enc_norm = m.patch_embed, m.enc_blocks, m.enc_norm
    def forward(self, img):
        x, pos = self.patch_embed(img)
        for blk in self.enc_blocks: x = blk(x, pos)
        return self.enc_norm(x)

class ASMKHead(nn.Module):
    def __init__(self, dim=MAST3R_ENC_DIM, cb=ASMK_DIM, tk=100):
        super().__init__()
        self.codebook = nn.Parameter(torch.empty(cb, dim))
        nn.init.xavier_uniform_(self.codebook)
        self.alpha = nn.Parameter(torch.tensor(3.0))
        self.top_k = tk
    def forward(self, f):
        fn, cn = F.normalize(f, dim=-1), F.normalize(self.codebook, dim=-1)
        s = torch.einsum('bnd,kd->bkn', fn, cn) * self.alpha
        k = min(self.top_k, s.shape[-1])
        tv, _ = torch.topk(s, k, dim=-1)
        return F.normalize(F.softmax(tv, dim=-1).sum(dim=-1), dim=-1)

asmk_backbone = MAST3RBackbone(MAST3R_PATH).to(device).eval()
asmk_head = ASMKHead().to(device)
ckpt = torch.load(ASMK_CKPT, **_TLK)
# Strip 'mast3r.' prefix — training saved full model, inference only needs encoder
bb_state = {k.replace('mast3r.', ''): v for k, v in ckpt['backbone_state'].items()
            if k.startswith('mast3r.patch_embed') or k.startswith('mast3r.enc_blocks') or k.startswith('mast3r.enc_norm')}
asmk_backbone.load_state_dict(bb_state, strict=True)
asmk_head.load_state_dict(ckpt['asmk_state'])
asmk_head.eval()
print(f'MAST3R-ASMK loaded (epoch {ckpt["epoch"]}). Dim: {ASMK_DIM}')
```

### 3c. MAST3R-SPoC (trained head on MAST3R backbone)
SPoC: Sum-Pooling + MLP projection → 256-dim global descriptor

```python
SPOC_IMG = 224
SPOC_DIM = 256
SPOC_CKPT = str(CHECKPOINT_DIR / 'mast3r_spoc' / 'best_model.pth')

class SPoCHead(nn.Module):
    def __init__(self, dim=MAST3R_ENC_DIM, hd=2048, od=256):
        super().__init__()
        self.projection = nn.Sequential(
            nn.Linear(dim, hd), nn.BatchNorm1d(hd),
            nn.ReLU(inplace=True), nn.Linear(hd, od))
    def forward(self, f):
        return F.normalize(self.projection(f.sum(dim=1)), dim=-1)

spoc_backbone = MAST3RBackbone(MAST3R_PATH).to(device).eval()
spoc_head = SPoCHead().to(device)
ckpt = torch.load(SPOC_CKPT, **_TLK)
bb_state = {k.replace('mast3r.', ''): v for k, v in ckpt['backbone_state'].items()
            if k.startswith('mast3r.patch_embed') or k.startswith('mast3r.enc_blocks') or k.startswith('mast3r.enc_norm')}
spoc_backbone.load_state_dict(bb_state, strict=True)
spoc_head.load_state_dict(ckpt['spoc_head_state'])
spoc_head.eval()
print(f'MAST3R-SPoC loaded (epoch {ckpt["epoch"]}). Dim: {SPOC_DIM}')
```

### 3d. ISC (ISC21 Descriptor Track 1st Place)
Source: `isc_feature_extractor` (EfficientNet-B3 + GeM, pretrained on copy detection)
Output: 256-dim L2-normalized descriptor

```python
from isc_feature_extractor import create_model

ISC_IMG = 384
ISC_DIM = 256

isc_model, isc_preprocessor = create_model(weight_name='isc_ft_v107', device=device)
isc_model.eval()
print(f'ISC (isc_ft_v107) loaded. Dim: {ISC_DIM}')
```

## 4. Transforms & Feature Extraction

```python
def make_transform(size):
    return transforms.Compose([
        transforms.Resize(int(size * 1.14)),
        transforms.CenterCrop(size),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ])

@torch.no_grad()
def extract_features_batch(model_or_tuple, model_name, paths, transform, batch_size, isc_preproc=None, **kwargs):
    """Generic batch feature extraction for a list of paths. Returns (N, D) tensor."""
    models = list(model_or_tuple) if isinstance(model_or_tuple, (tuple, list)) else [model_or_tuple]
    for m in models: m.eval()
    features = []
    for i in tqdm(range(0, len(paths), batch_size), desc=f'  {model_name}', leave=False):
        batch_paths = paths[i:i+batch_size]
        imgs = []
        for p in batch_paths:
            img = Image.open(p).convert('RGB')
            if isc_preproc:
                imgs.append(isc_preproc(img).unsqueeze(0))
            else:
                imgs.append(transform(img).unsqueeze(0))
        x = torch.cat(imgs, dim=0).to(device)
        if model_name == 'asmk':
            f = models[0](x); f = kwargs['asmk_head'](f)
        elif model_name == 'spoc':
            f = models[0](x); f = kwargs['spoc_head'](f)
        else:
            f = models[0](x)
        features.append(f.cpu())
    return torch.cat(features, dim=0)

# Extract features per dataset
print('Extracting features per dataset...')
t0 = time.time()

all_features = {}  # dataset_name -> {model_name: feature_tensor}

for ds_name, ds_images in tqdm(datasets.items(), desc='Datasets'):
    image_paths = [str(p) for p in ds_images]
    N_ds = len(image_paths)
    ds_out = OUTPUT_DIR / ds_name
    ds_out.mkdir(parents=True, exist_ok=True)
    
    print(f'\nDataset: {ds_name} ({N_ds} images)')
    
    feat_dict = {}
    feat_dict['dino'] = extract_features_batch(dino_model, 'dino', image_paths, make_transform(DINO_INPUT), 32)
    feat_dict['asmk'] = extract_features_batch((asmk_backbone,), 'asmk', image_paths, make_transform(ASMK_IMG), 8, asmk_head=asmk_head)
    feat_dict['spoc'] = extract_features_batch((spoc_backbone,), 'spoc', image_paths, make_transform(SPOC_IMG), 8, spoc_head=spoc_head)
    feat_dict['isc']  = extract_features_batch(isc_model, 'isc', image_paths, None, 32, isc_preproc=isc_preprocessor)
    
    # Save per-dataset features
    torch.save(feat_dict, ds_out / 'features.pt')
    all_features[ds_name] = feat_dict
    
    print(f'  DINO: {feat_dict["dino"].shape}, ASMK: {feat_dict["asmk"].shape}, '
          f'SPoC: {feat_dict["spoc"].shape}, ISC: {feat_dict["isc"].shape}')

print(f'\nFeature extraction done in {time.time()-t0:.0f}s')
```

## 5. Per-Dataset Shortlist (FPS-based ASMK + Top-K Union)

ASMK uses Farthest Point Sampling (n=10 seeds) + k=25 neighbors per seed — matching the champion's approach.
SPoC, DINOv2, ISC use simple top-K. Final shortlist = union of all four.

```python
def cos_sim(f):
    """Cosine similarity matrix (N, N)"""
    fn = F.normalize(f, dim=-1)
    return torch.mm(fn, fn.t())

def add_topk_to_shortlist(shortlist_sets, sim_matrix, k):
    """Add per-image top-k neighbors from one similarity matrix into a union shortlist."""
    N = sim_matrix.shape[0]
    if N <= 1:
        return shortlist_sets
    s = sim_matrix.clone()
    s.fill_diagonal_(-float('inf'))
    _, idxs = torch.topk(s, k=min(k, N - 1), dim=-1)
    for i in range(N):
        shortlist_sets[i].update(idxs[i].tolist())
    return shortlist_sets

# Generate per-dataset shortlists
print('Generating per-dataset shortlists...')

for ds_name, ds_images in tqdm(datasets.items(), desc='Shortlist'):
    ds_out = OUTPUT_DIR / ds_name
    feat = all_features[ds_name]
    image_names = [p.name for p in ds_images]
    N_ds = len(image_names)
    
    # Compute similarity matrices
    sims = {
        'dino': cos_sim(feat['dino']),
        'asmk': cos_sim(feat['asmk']),
        'spoc': cos_sim(feat['spoc']),
        'isc':  cos_sim(feat['isc']),
    }
    
    # Save per-model similarity matrices
    sims_np = {k: v.cpu().numpy() for k, v in sims.items()}
    np.savez(ds_out / 'per_model_similarity.npz', **sims_np)
    
    # Per-model top-K union: DINOv2 + MASt3R-ASMK + MASt3R-SPoC + ISC.
    shortlist_sets = {i: set() for i in range(N_ds)}
    for key in ['asmk', 'spoc', 'dino', 'isc']:
        add_topk_to_shortlist(shortlist_sets, sims[key], PER_MODEL_TOPK[key])
    
    # Build shortlist JSON
    shortlist = {}
    for i in range(N_ds):
        name = image_names[i]
        cands = []
        for j in shortlist_sets[i]:
            if i == j:
                continue
            nbr = image_names[j]
            # Use max similarity across models as score
            score = max(sims[k][i, j].item() for k in sims)
            cands.append({'idx': j, 'name': nbr, 'score': float(score)})
        cands.sort(key=lambda x: x['score'], reverse=True)
        shortlist[name] = cands
    
    total = sum(len(v) for v in shortlist.values())
    print(f'  {ds_name}: {total} pairs, avg {total/N_ds:.1f}/image')
    
    # Save
    out = {k: [{'idx': c['idx'], 'name': c['name'], 'score': c['score']} for c in v]
           for k, v in shortlist.items()}
    with open(ds_out / 'shortlist.json', 'w') as f:
        json.dump(out, f, indent=2)

print(f'\nAll shortlists saved to {OUTPUT_DIR}/<dataset>/shortlist.json')
```

## 6. Similarity Visualization

```python
# Visualize similarity for the first dataset (or all if few)
first_ds = list(datasets.keys())[0]
feat = all_features[first_ds]
sims_viz = {
    'dino': cos_sim(feat['dino']),
    'asmk': cos_sim(feat['asmk']),
    'spoc': cos_sim(feat['spoc']),
    'isc':  cos_sim(feat['isc']),
}

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, (key, sim) in zip(axes, sims_viz.items()):
    n = min(100, sim.shape[0])
    im = ax.imshow(sim[:n, :n].numpy(), cmap='viridis', vmin=0, vmax=1)
    ax.set_title(f'{key} (top-{PER_MODEL_TOPK[key]}) [{first_ds}]')
    plt.colorbar(im, ax=ax, fraction=0.046)
plt.suptitle(f'Per-Model Similarity — {first_ds} (first 100)', fontsize=14)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'similarity.png', dpi=150)
plt.show()
```

## 9. Summary

```python
print('='*60)
print('Retrieval Complete')
print('='*60)
print(f'Datasets: {len(datasets)}')
for ds_name, ds_images in datasets.items():
    ds_out = OUTPUT_DIR / ds_name
    sl_path = ds_out / 'shortlist.json'
    if sl_path.exists():
        with open(sl_path) as f:
            sl = json.load(f)
        total = sum(len(v) for v in sl.values())
        print(f'  {ds_name}: {len(ds_images)} images, {total} pairs (avg {total/len(ds_images):.1f}/image)')
print(f'\nPer-model settings:')
print(f'  ASMK: FPS n={ASMK_FPS_N}, k={PER_MODEL_TOPK["asmk"]} (FPS-based)')
print(f'  SPoC: topk={PER_MODEL_TOPK["spoc"]}')
print(f'  DINOv2: topk={PER_MODEL_TOPK["dino"]} (MAC pooling, 768-dim)')
print(f'  ISC: topk={PER_MODEL_TOPK["isc"]}')
print(f'\nOutput: {OUTPUT_DIR}/<dataset>/')
print(f'  shortlist.json  -- candidate image pairs per dataset')
print(f'  image_paths.json -- name->path mapping per dataset')
print(f'  features.pt      -- feature tensors per dataset')
print(f'\nNext -> 02_mast3r_matching.ipynb (hybrid MASt3R matching)')
```
